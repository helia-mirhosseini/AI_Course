```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split


scaler = StandardScaler()
y = df['target']
x = df.drop("target", axis=1)

x_train,x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state=42)
```

```python

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

model  = LogisticRegression(max_iter=1000)
model.fit(x_train_scaled, y_train)
y_pred = model.predict(x_test_scaled)

print("the accuracy of the model is:", accuracy_score(y_pred, y_test))
print(classification_report(y_pred, y_test))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm_norm = cm.astype('float')/cm.sum(axis=1)[:,np.newaxis]
sns.heatmap(cm_norm, annot=True, fmt=".2f", cmap='Blues')
plt.title("Normalized Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_scaled, y_train)
knn_pred= knn.predict(x_test_scaled)
print("the accuracy for KNN algorithm is:", accuracy_score(knn_pred, y_test))
print(classification_report(knn_pred, y_test))
```

```python
cm  = confusion_matrix(knn_pred, y_test)
cm_norm = cm.astype('float')/cm.sum(axis=1)[:,np.newaxis]
sns.heatmap(cm_norm, annot=True, fmt=".2f", cmap='Blues')
plt.title("Normalized Confusion Matrix for KNN algorithm")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
from sklearn.svm import SVC

svc = SVC(kernel='rbf', C = 1.0, probability= True, gamma= 'scale')
svc.fit(x_train_scaled, y_train)
svc_pred = svc.predict(x_test_scaled)
print("the accuracy for SVC algorithm is:", accuracy_score(svc_pred, y_test))
print(classification_report(svc_pred, y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier

decission_tree = DecisionTreeClassifier(max_depth=5)
decission_tree.fit(x_train_scaled, y_train)
decission_tree_pred = decission_tree.predict(x_test_scaled)
print("the accuracy for KNN algorithm is:", accuracy_score(decission_tree_pred, y_test))
print(classification_report(decission_tree_pred, y_test))
```

```python

from sklearn.ensemble import BaggingClassifier

base_tree = DecisionTreeClassifier(max_depth=4, random_state=42)

bagging = BaggingClassifier(
    estimator=base_tree,       # base learner
    n_estimators=50,           # number of trees
    max_samples=0.8,           # 80% of training samples per tree
    bootstrap=True,            # sampling with replacement
    random_state=42
)

bagging.fit(x_train_scaled, y_train)
bagging_predict = bagging.predict(x_test_scaled)

print("the accuracy for KNN algorithm is:", accuracy_score(bagging_predict, y_test))
print(classification_report(bagging_predict, y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier

random  = RandomForestClassifier(n_estimators=100)
random.fit(x_train_scaled, y_train)
random_predict = random.predict(x_test_scaled)
print("the accuracy for KNN algorithm is:", accuracy_score(random_predict, y_test))
print(classification_report(random_predict, y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gradient  = GradientBoostingClassifier(n_estimators=100)
gradient.fit(x_train_scaled, y_train)
gradient_predict = gradient.predict(x_test_scaled)
print("the accuracy for KNN algorithm is:", accuracy_score(gradient_predict, y_test))
print(classification_report(gradient_predict, y_test))
```
