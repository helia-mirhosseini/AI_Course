<h2  align = "right" class = "rtl">
<strong>خواندن دیتاست</strong></h2>
<p class = "rtl">
بعد از خواندن دیتاست با استقاده از دستور های کتابخانه<span class = "ltr"><code>pandas</code></span> نیاز داریم که از بین ستون های دیتاست ستون مربوط به ویژگی های هدف با همان <span class ="ltr"> target</span> را انتخاب کنیم. این ستون باید توسط مدل ما پیش بینی شود. این ستون را معادل سپرده یا همان <span class = "ltr"> deposit</span> قرار میدهیم. 
</p>
<p class = "rtl">
در این کد هدف این است که بفهمیم مشتری فعلی بانک به خدمات آینده بانک پاسخ مثبت میدهد یا خیر. با بررسی اینکه مشتری ما در حال حاضر در بانک حساب دارد یا خیز میتوانیم پیش بینی کنیم که چه واکنشی به پیشمهاد های آینده بانک خواهد داد. به همین علت ستون <span class = "ltr">deposit</span> یه عنوان ویژگی هدف انتخاب شده. 
</p>
<p class = "rtl">
در مرحله بعدی باید مقادیر این ستون را به باینری تبدیل کنیم. تا مدل ما بتواند آن را متوجه شود. بعد از تبدیل مقادیر ستون باید دیتاست مربوط به هدف را از دیتاست <span class = "ltr">X</span>جدا کنیم. 
</p>

```python
# Load the dataset
df = pd.read_csv("bank.csv")  

# Identify the correct label column
print(df.columns) 
target_col = 'deposit'

# Convert target to binary
df[target_col] = df[target_col].map({'yes': 1, 'no': 0})

# Separate features and labels
y = df[target_col]
X_raw = df.drop(columns=[target_col])

```

<h2 align  = "right" class = "ltr">
<strong>one-hot encoding</strong>
</h2>
<p class = "rtl">
در این قسمت میخواهیم روش <span class = "ltr">one-hot encoding</span> را بررسی کنیم. این روش را قبلا در نوت بوک مربوط به اکتشاف داده دیدیم.این روش زمانی استفاده میشود که دیتاست ما مقادیر <span class = "ltr">string</span> زیادی داشته باشد. ولی باید این روش را با دقت زیادی اجرا کرد؛ چراکه مقدار ستون های دیتاست ما را به طرز چشمگیری افزایش میدهد. 
</p>
<p class = "rtl">
برای درک بهتر عملکرد <span class = "ltr"> one-hot encoding </span> یکبار دیتاست اولیه را پرینت کرده و با دیتاست <span class = "ltr">X_onehot </span> مقایسه کنید. به عنوان مثال به ستوت <span class = "ltr">job </span> نگاه کنید. این ستون شامل شغل مشتری های بانک بوده و تمام مقادیر آن از نوع <span class = "ltr"> string </span> میباشد. کاری که روش <span class = "ltr"> one-hot encoding </span> انجام میدهد این است که برای هر شغل ستونی با نام مربوطه ایجاد میکند و مقدار آن را با توجه به دیتاست اصلی <span class = "ltr"> False </span> یا <span class = "ltr"> True </span> قرار میدهد.
</p>
<p class = "rtl">
ستون <span class = "ltr"> 	job_blue-collar </span> را نگاه کنید. این ستون تنها برای کسانی که شغل های خدماتی دارند <span class = "ltr"> True </span> خواهد بود و برای بقیه مشتری های بانک <span class = "ltr"> False </span> میباشد.
</p>
<p class = "rtl">
برای پیدا کردن تاثیر هر کدام از فیچر ها از تابع<span class= "ltr"> coef </span>استفاده میکنیم. این تابع وزن های اطلاق شده توسط مدل به
هر فیچر را به ما نشان میدهد. با  <span class= "ltr"> sort</span>کردن این مقادیر میتوان فهمید کدام یک از فیچر ها بیشترین تاثیر را در
پیش بینی داشته اند. در ادامه 10 ویژگی برتر را پرینت میکنیم.
</p>

```python
# ----------------------METHOD 1: ONE-HOT ENCODING------------------------------
X_onehot = pd.get_dummies(X_raw, drop_first=True)
X1_train, X1_test, y1_train, y1_test = train_test_split(X_onehot, y, test_size=0.2, random_state=42)

model1 = LogisticRegression(max_iter=1000)
model1.fit(X1_train, y1_train)
y1_pred = model1.predict(X1_test)
acc1 = accuracy_score(y1_test, y1_pred)

# Analyze feature importance
feature_importance = pd.Series(model1.coef_[0], index=X1_train.columns).sort_values(key=abs, ascending=False)
print("\nTop 10 Features Influencing Customer Decision(with One-Hot Encoding):")
print(feature_importance.head(10))

X_onehot
```

<h2 align = "right" class = "ltr">
<strong>label encoding</strong>
</h2>
<p class = "rtl">
در این قسمت میخواهیم روش <span class = "ltr"> label encoding</span> را بررسی کنیم. در این یاز هم مقادیر <span class= "ltr">string</span> به مقادیر عددی مپ خواهند شد ولی شیوه عملکرد ما با روش <span class= "ltr">one-hot encoding</span> متقاوت است. 
</p>
<p class = "rtl">
در این روش ما تعداد <span class= "ltr">value</span> های مقاوت در هر ستون که محتوای <span class= "ltr">string</span> دارد را میشماریم و به جای آنها عدد قرار میدهیم. ستون شغل ها را به عنوان نمونه نگاه کنید. این ستون شامل 12 شغل متفاوت است. زمانی که از این روش استقاده میکنیم؛ به جای اضافه شدن 12 ستون متفاوت به دیتاست، مقادیر ستون به اعداد 0 تا 11 مپ خواهد شد. 
</p>
<p class = "rtl">
این روش زمانی اهمیت پیدا میکند که نمیخواهیم ستون تعداد ستون های دیتاست بیش از حد زیاد شود. ولی در کل زمانی که دیتاست ما مقادیر <span class= "ltr">string</span> زیادی دارد بهتر است این مقادیر ابتدا با مقادیر عددی جایگزین شوند(روش های آن در ویدئو<span class= "ltr">data exploration </span> توضیح داده شد.) و بعد با استفاده از یکی از این دو تکنیک ستون های باقی مانده که شامل مقادیر <span class= "ltr">string</span> هستند را جایگزین کنیم.
</p>

```python
# --------------------METHOD 2: LABEL ENCODING---------------------
X_label = X_raw.copy()
label_encoders = {}

for col in X_label.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X_label[col] = le.fit_transform(X_label[col])
    label_encoders[col] = le  # Save encoders if you want to inverse later

X2_train, X2_test, y2_train, y2_test = train_test_split(X_label, y, test_size=0.2, random_state=42)

model2 = LogisticRegression(max_iter=1000)
model2.fit(X2_train, y2_train)
y2_pred = model2.predict(X2_test)
acc2 = accuracy_score(y2_test, y2_pred)

# Analyze feature importance
feature_importance = pd.Series(model2.coef_[0], index=X2_train.columns).sort_values(key=abs, ascending=False)
print("\nTop 10 Features Influencing Customer Decision(with Label Encoding):")
print(feature_importance.head(10))

X2_train['job'].value_counts()
```

اما این دو روش چه تاثیری روی دقت مدل دارند؟
</strong>
</h2>
<p class = "rtl">
دلیل اینکه با روش اول نتایج بهتری گرفتیم این است که  <span class = "ltr"> label encoding </span> روابط عددی اشتباه ایجاد میکند. در
 ،<span class = "ltr">label encoding</span> مقادیر متنی (مثلاً شغل ها) به اعداد تبدیل میشوند و مدل اینگونه برداشت میکند که این شغل ها
دارای ترتیب و فاصله عددی هستند، اما در واقع این اعداد صرفاً برچسب هستند و ترتیب آنها هیچ مفهومی ندارد .
بنابراین مدل دچار خطا میشود.
</p>
<p class = "rtl">
ولی در  <span class = "ltr">one-hot encoding</span>ما متغیر ها را کاملا مستقل از یکدیگر تعریف میکنیم و میفهمیم که هیچ ربطی به همدیگر
ندارند.
بنابراین برای مدل های خطی مثل  <span class = "ltr">logistic regression</span>از  <span class = "ltr">one-hot encoding</span>استفاده میکنیم چراکه متغیر ها
فاصله ای با یکدیگر ندارند و اینگونه مدل ما اشتباه یاد نمیگیرد.
</p>

```python
# Comparing the accuracy 
from sklearn.metrics import classification_report
print(f"Accuracy with One-Hot Encoding: {acc1:.4f}")
print(f"Accuracy with Label Encoding:   {acc2:.4f}")
print(classification_report(y1_test, y1_pred, digits=3))
```
