<h2 align="right" class="rtl" style="color:#FFBF00;">
  تعریف مسأله: طبقه‌بندی گل‌ها با استفاده از درخت تصمیم
</h2>

<p class="rtl">
  در این تمرین، هدف ما ایجاد یک مدل درخت تصمیم است که بتواند انواع گل‌های <span class="ltr">Iris</span> را با استفاده از ویژگی‌های مختلف آن‌ها (مثل طول و عرض گلبرگ و درپوش) طبقه‌بندی کند. این داده‌ها از یک مجموعه داده استاندارد به نام <span class="ltr">Iris</span> که در کتابخانه‌ی <span class="ltr">sklearn</span> وجود دارد، بارگذاری می‌شود. 
  هدف نهایی ما این است که مدل درخت تصمیم را آموزش دهیم و سپس آن را ارزیابی کنیم تا عملکرد مدل در طبقه‌بندی گل‌ها را بسنجیم.
</p>

<p class="rtl">
  برای این کار، از داده‌هایی که ویژگی‌هایی مانند طول و عرض گلبرگ و درپوش برای هر گل را دارند استفاده خواهیم کرد تا مدل درخت تصمیم یاد بگیرد که گل‌ها را بر اساس این ویژگی‌ها به سه دسته‌ی مختلف طبقه‌بندی کند. دسته‌ها عبارتند از:
</p>

<ol dir="rtl" class="rtl" style="padding-right:1.25em">
  <li><strong>Iris-setosa:</strong> نوعی از گل‌ها که ویژگی‌های خاص خود را دارند.</li>
  <li><strong>Iris-versicolor:</strong> نوعی دیگر از گل‌ها که دارای ویژگی‌های متفاوت است.</li>
  <li><strong>Iris-virginica:</strong> نوع سوم از گل‌ها که با ویژگی‌های خاص خود شبیه به دو نوع دیگر است.</li>
</ol>

<p class="rtl">
  در این پروژه، الگوریتم درخت تصمیم به ما کمک می‌کند تا با استفاده از داده‌های موجود، یک مدل پیش‌بینی ایجاد کنیم که بتواند برای گل‌های جدیدی که ویژگی‌های آن‌ها به مدل داده می‌شود، نوع گل را پیش‌بینی کند.
</p>

<h2 align="right" class="rtl" style="color:    color: #FFBF00;">
  بارگذاری کتابخانه‌های موردنیاز
</h2>

<p class="rtl">
  در این سلول، کتابخانه‌های لازم برای اجرای مدل درخت تصمیم بارگذاری می‌شوند:
</p>

<ol dir="rtl" class="rtl" style="padding-right:1.25em">
  <li><strong>train_test_split:</strong> از کتابخانه‌ی <span class="ltr">sklearn.model_selection</span>، داده‌ها به دو بخش آموزش و آزمون تقسیم می‌شود تا مدل آموزش داده شود و سپس ارزیابی گردد.</li>
  <li><strong>DecisionTreeClassifier:</strong> از کتابخانه‌ی <span class="ltr">sklearn.tree</span>، الگوریتم درخت تصمیم برای دسته‌بندی داده‌ها استفاده می‌شود.</li>
  <li><strong>accuracy_score:</strong> برای ارزیابی دقت مدل استفاده می‌شود که نشان می‌دهد مدل چقدر درست پیش‌بینی کرده است.</li>
  <li><strong>classification_report:</strong> گزارشی دقیق از عملکرد مدل ارائه می‌دهد که شامل معیارهایی مانند دقت، یادآوری و امتیاز F1 است.</li>
  <li><strong>confusion_matrix:</strong> ماتریس اغتشاش را برای ارزیابی عملکرد مدل تولید می‌کند که نشان می‌دهد مدل چه تعداد پیش‌بینی درست و غلط داشته است.</li>
  <li><strong>matplotlib و seaborn:</strong> برای ترسیم نمودارها و تجسم داده‌ها استفاده می‌شوند.</li>
</ol>


```python
# Import necessary libraries
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
```

<h2 align="right" class="rtl" style="color:    color: #FFBF00;">
  بارگذاری مجموعه داده و تقسیم داده‌ها
</h2>

<p class="rtl">
  در این سلول، داده‌های مربوط به گل‌ها (مجموعه داده <span class="ltr">Iris</span>) بارگذاری می‌شود و سپس به دو مجموعه آموزشی و آزمایشی تقسیم می‌شود:
</p>

<ol dir="rtl" class="rtl" style="padding-right:1.25em">
  <li><strong>بارگذاری داده‌های Iris:</strong> با استفاده از تابع <span class="ltr">load_iris()</span> داده‌های مجموعه <span class="ltr">Iris</span> که شامل ویژگی‌های مختلف گل‌ها و نوع هر گل است، بارگذاری می‌شود. این داده‌ها شامل ۴ ویژگی هستند: طول و عرض گلبرگ و درپوش (به ترتیب <span class="ltr">Sepal length, Sepal width, Petal length, Petal width</span>) که برای شناسایی نوع گل‌ها استفاده می‌شوند.</li>
  
  <li><strong>ویژگی‌ها و هدف:</strong> 
    <ul>
      <li><strong>X:</strong> ویژگی‌ها که در اینجا همان طول و عرض گلبرگ و درپوش هستند.</li>
      <li><strong>y:</strong> هدف یا برچسب که نوع گل را نشان می‌دهد (سه نوع گل: <span class="ltr">Setosa, Versicolor, Virginica</span>).</li>
    </ul>
  </li>
  
  <li><strong>تقسیم داده‌ها به مجموعه‌های آموزشی و آزمایشی:</strong> 
    <ul>
      <li>با استفاده از تابع <span class="ltr">train_test_split()</span> داده‌ها به دو مجموعه آموزشی (۷۰٪) و آزمایشی (۳۰٪) تقسیم می‌شود. این کار برای آموزش مدل و سپس ارزیابی عملکرد آن بر روی داده‌های جدید انجام می‌شود.</li>
      <li><strong>test_size=0.3:</strong> این پارامتر نشان می‌دهد که ۳۰٪ از داده‌ها برای آزمایش مدل استفاده می‌شوند و باقی‌مانده برای آموزش مدل به کار می‌روند.</li>
      <li><strong>random_state=42:</strong> این پارامتر برای تولید نتایج قابل تکرار در تقسیم داده‌ها استفاده می‌شود. این مقدار اطمینان می‌دهد که تقسیم‌بندی داده‌ها همیشه به همان شکل انجام شود.</li>
    </ul>
  </li>
</ol>

<p class="rtl">
  در نهایت، داده‌های <span class="ltr">X_train</span> و <span class="ltr">y_train</span> برای آموزش مدل استفاده می‌شوند و <span class="ltr">X_test</span> و <span class="ltr">y_test</span> برای ارزیابی دقت مدل مورد استفاده قرار خواهند گرفت.
</p>


```python
# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features: Sepal length, Sepal width, Petal length, Petal width
y = iris.target  # Target: Species (Setosa, Versicolor, Virginica)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

<h2 align="right" class="rtl" style="color: #FFBF00;">
  نحوه عملکرد درخت تصمیم
</h2>

<p class="rtl">
  درخت تصمیم یک الگوریتم یادگیری ماشین است که برای حل مسائل دسته‌بندی و رگرسیون استفاده می‌شود. این الگوریتم با استفاده از داده‌های آموزشی، یک مدل درختی ایجاد می‌کند که به‌صورت سلسله‌مراتبی تصمیم‌گیری می‌کند. هر گره درخت نمایانگر یک ویژگی از داده‌هاست و هر برگ نشان‌دهنده یک پیش‌بینی یا تصمیم است.
</p>

<p class="rtl">
  درخت تصمیم به‌طور خاص از طریق فرآیندهای زیر عمل می‌کند:
</p>

<ol dir="rtl" class="rtl" style="padding-right:1.25em">
  <li><strong>تقسیم داده‌ها:</strong> 
    <ul>
      <li>در هر گره درخت، داده‌ها به دو یا چند گروه تقسیم می‌شوند. این تقسیم‌بندی بر اساس بهترین ویژگی‌ها و آستانه‌هایی است که داده‌ها را به‌طور مؤثر به کلاس‌های مختلف (در مسائل دسته‌بندی) یا مقادیر مختلف (در مسائل رگرسیون) تقسیم می‌کند.</li>
    </ul>
  </li>
  
  <li><strong>انتخاب بهترین ویژگی برای تقسیم:</strong> 
    <ul>
      <li>درخت تصمیم برای تقسیم داده‌ها از معیارهای مختلفی مانند <strong>انحراف معیار (Gini Impurity)</strong> یا <strong>آنتروپی</strong> (در الگوریتم <span class="ltr">CART</span>) استفاده می‌کند. هدف این است که داده‌ها به‌گونه‌ای تقسیم شوند که هر گروه تا حد ممکن مشابه یکدیگر باشد.</li>
      <li><strong>Gini Impurity:</strong> این معیار نشان می‌دهد که چقدر احتمال دارد که یک نمونه به اشتباه به کلاس دیگری تخصیص یابد. هرچه مقدار Gini پایین‌تر باشد، گروه‌های ایجاد شده از تقسیم‌بندی بهتر هستند.</li>
      <li><strong>آنتروپی:</strong> این معیار به میزان عدم‌اطمینان یا بی‌نظمی در یک مجموعه داده اشاره دارد. هدف کاهش آنتروپی در هر تقسیم است.</li>
    </ul>
  </li>
  
  <li><strong>ساخت درخت:</strong> 
    <ul>
      <li>این فرآیند ادامه می‌یابد تا زمانی که تمام داده‌ها به‌طور کامل طبقه‌بندی شوند یا معیارهای خاصی مانند عمق درخت یا تعداد نمونه‌ها به حد خاصی برسد.</li>
    </ul>
  </li>
  
  <li><strong>پیش‌بینی:</strong> 
    <ul>
      <li>پس از ساخت درخت، برای پیش‌بینی یک نمونه جدید، داده‌ها از ریشه درخت شروع به حرکت می‌کنند و در هر گره به سمت یکی از دو یا چند مسیر ممکن حرکت می‌کنند. این مسیرها بر اساس ویژگی‌های داده‌ها انتخاب می‌شوند. این فرآیند تا رسیدن به یک برگ ادامه می‌یابد، که در آن پیش‌بینی نهایی داده می‌شود.</li>
    </ul>
  </li>
</ol>

<p class="rtl">
  به‌طور خلاصه، درخت تصمیم یک مدل سلسله‌مراتبی است که با استفاده از تقسیم داده‌ها در گره‌ها به پیش‌بینی‌هایی می‌پردازد. این مدل برای مسائل دسته‌بندی و رگرسیون بسیار مناسب است زیرا تفسیر آن ساده و قابل‌فهم است.
</p>

<h2 align="right" class="rtl" style="color: #FFBF00;">
  آموزش مدل درخت تصمیم و پیش‌بینی
</h2>

<p class="rtl">
  در این سلول، مدل درخت تصمیم ایجاد و آموزش داده می‌شود، سپس پیش‌بینی‌هایی برای داده‌های آزمایشی انجام می‌گردد:
</p>

<ol dir="rtl" class="rtl" style="padding-right:1.25em">
  <li><strong>ایجاد مدل درخت تصمیم:</strong> 
    <ul>
      <li>با استفاده از کلاس <span class="ltr">DecisionTreeClassifier</span> یک مدل درخت تصمیم ایجاد می‌شود. این مدل برای دسته‌بندی داده‌ها به‌کار می‌رود و به‌طور خودکار قوانین را از داده‌های آموزشی استخراج می‌کند.</li>
      <li>پارامتر <span class="ltr">random_state=42</span> برای تضمین تکرارپذیری نتایج استفاده می‌شود. این باعث می‌شود که مدل همیشه به همان روش آموزش داده شود.</li>
    </ul>
  </li>

  <li><strong>آموزش مدل:</strong>
    <ul>
      <li>با استفاده از متد <span class="ltr">fit()</span> مدل درخت تصمیم روی داده‌های آموزشی <span class="ltr">X_train</span> و <span class="ltr">y_train</span> آموزش داده می‌شود. در این مرحله، مدل سعی می‌کند الگوها و روابط میان ویژگی‌ها و برچسب‌ها را پیدا کند.</li>
    </ul>
  </li>

  <li><strong>پیش‌بینی:</strong> 
    <ul>
      <li>پس از آموزش مدل، برای پیش‌بینی برچسب گل‌ها در داده‌های آزمایشی از متد <span class="ltr">predict()</span> استفاده می‌شود. این متد برای داده‌های <span class="ltr">X_test</span> پیش‌بینی‌هایی انجام می‌دهد و نتایج پیش‌بینی در متغیر <span class="ltr">y_pred</span> ذخیره می‌شود.</li>
    </ul>
  </li>
</ol>

<p class="rtl">
  به‌طور کلی، این سلول وظیفه دارد تا مدل را روی داده‌های آموزشی آموزش دهد و سپس از آن برای پیش‌بینی نتایج برای داده‌های جدید (داده‌های آزمایشی) استفاده کند.
</p>


```python
# Initialize and train the Decision Tree model
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)
```

<h2 align="right" class="rtl" style="color: #FFBF00;">
  ارزیابی مدل
</h2>

<p class="rtl">
  در این سلول، عملکرد مدل درخت تصمیم ارزیابی می‌شود. دو معیار اصلی برای ارزیابی مدل استفاده می‌شود: دقت مدل و گزارش طبقه‌بندی.
</p>

<ol dir="rtl" class="rtl" style="padding-right:1.25em">
  <li><strong>دقت مدل:</strong> 
    <ul>
      <li>برای ارزیابی دقت مدل از متد <span class="ltr">accuracy_score()</span> استفاده می‌شود. این متد تعداد پیش‌بینی‌های درست را تقسیم بر تعداد کل نمونه‌های داده آزمایشی (داده‌های <span class="ltr">X_test</span>) محاسبه می‌کند. نتیجه این محاسبه به‌صورت درصد نمایش داده می‌شود.</li>
      <li>در اینجا، دقت مدل با فرمت درصد به نمایش درمی‌آید. عبارت <span class="ltr">accuracy * 100</span> برای تبدیل دقت به درصد استفاده می‌شود.</li>
    </ul>
  </li>
  
  <li><strong>گزارش طبقه‌بندی:</strong> 
    <ul>
      <li>برای ارزیابی بیشتر عملکرد مدل، از متد <span class="ltr">classification_report()</span> استفاده می‌شود. این متد گزارشی دقیق از معیارهای مختلفی مانند دقت (precision)، یادآوری (recall) و امتیاز F1 برای هر کلاس (در اینجا برای هر نوع گل) ارائه می‌دهد.</li>
      <li><strong>دقت (Precision):</strong> نشان‌دهنده درصد پیش‌بینی‌های درست در بین نمونه‌هایی است که به‌عنوان یک کلاس خاص (مثلاً یک نوع گل) شناسایی شده‌اند.</li>
      <li><strong>یادآوری (Recall):</strong> نشان‌دهنده درصد نمونه‌های واقعی از یک کلاس خاص است که به‌درستی توسط مدل شناسایی شده‌اند.</li>
      <li><strong>امتیاز F1:</strong> میانگین هماهنگی دقت و یادآوری است که عملکرد مدل را به‌طور جامع‌تر ارزیابی می‌کند.</li>
      <li>در نهایت، این گزارش برای هر یک از سه کلاس موجود در مجموعه داده (Setosa، Versicolor، Virginica) به‌طور جداگانه تولید می‌شود.</li>
    </ul>
  </li>
</ol>

<p class="rtl">
  این ارزیابی‌ها به ما کمک می‌کند تا عملکرد مدل را در شناسایی دقیق انواع مختلف گل‌ها بررسی کنیم و همچنین ببینیم که مدل در هر کدام از کلاس‌ها چگونه عمل کرده است.
</p>


```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Print the classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```

<h2 align="right" class="rtl" style="color: #FFBF00;">
  ترسیم ماتریس اغتشاش (<span class="ltr">Confusion Matrix</span>)
</h2>

<p class="rtl">
  در این سلول، عملکرد مدل با استفاده از <strong>ماتریس اغتشاش</strong> به‌صورت تصویری نمایش داده می‌شود. این ماتریس یکی از ابزارهای اصلی برای بررسی میزان دقت و نوع خطاهای مدل در مسائل دسته‌بندی است.
</p>
