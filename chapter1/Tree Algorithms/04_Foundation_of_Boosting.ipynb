<div class="note-box">
    <h2>4. Learning from Mistakes: Boosting</h2>
    <p>
        In the previous notebook, we used <strong>Bagging</strong> (Random Forest). We trained many trees independently (in parallel) and averaged their results. This reduced <em>Variance</em>.
    </p>
    <p>
        Now, we switch to <strong>Boosting</strong>. This is a <strong>Sequential</strong> process.
    </p>
    <ul>
        <li><strong>Bagging (Parallel):</strong> 100 students take a test independently. We average their scores.</li>
        <li><strong>Boosting (Sequential):</strong> Student 1 takes a test. Student 2 looks at Student 1's mistakes and studies <em>only those topics</em>. Student 3 looks at Student 2's mistakes, and so on.</li>
    </ul>
    <p>
        Boosting reduces <strong>Bias</strong> (it makes the model smarter), essentially converting many weak learners (shallow trees) into one strong learner.
    </p>
</div>

<div class="note-box">
    <h2>The Two Families of Boosting</h2>
    <h3>1. AdaBoost (Adaptive Boosting)</h3>
    <p>
        The classic approach. It works by manipulating <strong>sample weights</strong>.
    </p>
    <ul>
        <li>Train Tree 1.</li>
        <li>Identify the samples it got wrong.</li>
        <li><strong>Increase the weight</strong> of those "hard" samples so the next tree <em>must</em> pay attention to them.</li>
    </ul>

<h3>2. Gradient Boosting (GBM)</h3>
    <p>
        The mathematical approach. It works by manipulating the <strong>Target (Y)</strong>.
    </p>
    <ul>
        <li>Train Tree 1 to predict the target <span class="code-inline">y</span>.</li>
        <li>Calculate the error (Residual): <span class="code-inline">r1 = y - prediction</span>.</li>
        <li>Train Tree 2 to predict <strong>r1</strong> (the error), not y!</li>
        <li>The final prediction is <span class="code-inline">Tree1 + Tree2 + Tree3...</span></li>
    </ul>
</div>

<div class="note-box">
    <h2>Exercise: DIY Gradient Boosting</h2>
    <p>
        To truly understand GBM, we will build one from scratch using 3 simple Decision Trees.
    </p>
    <p>
        <strong>The Goal:</strong> Fit a curved line (Quadratic data) using straight lines (Decision Trees).
    </p>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# 1. Create Data (A simple curve with some noise)
np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100) # Quadratic y = 3x^2 + noise

# 2. Train Tree 1 on the raw data
tree_reg1 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg1.fit(X, y)

# 3. Calculate the Error (Residual) of Tree 1
# y2 is the "mistake" Tree 1 made
y2 = y - tree_reg1.predict(X)

# 4. Train Tree 2 on the Residuals (y2)
tree_reg2 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg2.fit(X, y2)

# 5. Calculate the Error of Tree 2
y3 = y2 - tree_reg2.predict(X)

# 6. Train Tree 3 on the new Residuals (y3)
tree_reg3 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg3.fit(X, y3)

# --- Visualize the Addition of Trees ---
def plot_predictions(regressors, X, y, axes, label=None, style="r-", data_style="b.", data_label=None):
    x1 = np.linspace(-0.5, 0.5, 500)
    y_pred = sum(reg.predict(x1.reshape(-1, 1)) for reg in regressors)
    axes.plot(X[:, 0], y, data_style, label=data_label)
    axes.plot(x1, y_pred, style, linewidth=2, label=label)
    if label or data_label:
        axes.legend(loc="upper center", fontsize=10)

plt.figure(figsize=(11,11))

# Row 1: The Residuals (What the tree is learning)
plt.subplot(321)
plot_predictions([tree_reg1], X, y, plt.gca(), label="Tree 1 Prediction", data_label="Training Data")
plt.title("Tree 1: Fits Data")

plt.subplot(323)
plot_predictions([tree_reg2], X, y2, plt.gca(), label="Tree 2 Prediction", data_style="k+", data_label="Residuals (y-h1)")
plt.title("Tree 2: Fits Tree 1's Errors")

plt.subplot(325)
plot_predictions([tree_reg3], X, y3, plt.gca(), label="Tree 3 Prediction", data_style="k+", data_label="Residuals (y2-h2)")
plt.title("Tree 3: Fits Tree 2's Errors")

# Row 2: The Ensemble (The Sum of Trees)
plt.subplot(322)
plot_predictions([tree_reg1], X, y, plt.gca(), label="h(x) = h1(x)", data_label="Training Data")
plt.title("Ensemble: Tree 1")

plt.subplot(324)
plot_predictions([tree_reg1, tree_reg2], X, y, plt.gca(), label="h(x) = h1(x) + h2(x)")
plt.title("Ensemble: Tree 1 + Tree 2")

plt.subplot(326)
plot_predictions([tree_reg1, tree_reg2, tree_reg3], X, y, plt.gca(), label="h(x) = h1(x) + h2(x) + h3(x)")
plt.title("Ensemble: Tree 1 + Tree 2 + Tree 3")

plt.tight_layout()
plt.show()
```

<div class="note-box">
    <h2>Interpreting the Plots</h2>
    <p>
        <strong>Left Column (Individual Learners):</strong>
    </p>
    <ul>
        <li><strong>Top:</strong> Tree 1 tries to fit the curve but creates a blocky, stair-step shape.</li>
        <li><strong>Middle:</strong> Tree 2 isn't looking at the curve! It is looking at the <strong>errors</strong> (residuals) from Tree 1. It finds where Tree 1 was too low or too high.</li>
        <li><strong>Bottom:</strong> Tree 3 cleans up the small errors left by Tree 2.</li>
    </ul>

<p>
        <strong>Right Column (The Ensemble):</strong>
    </p>
    <ul>
        <li>As we add the trees together (<span class="code-inline">Pred = T1 + T2 + T3</span>), the prediction gets closer and closer to the actual curve.</li>
        <li>This is <strong>Gradient Boosting</strong> in a nutshell.</li>
    </ul>
</div>

<div class="note-box">
    <h2>4. Using Scikit-Learn</h2>
    <p>
        We don't need to do this manually. SKLearn gives us <span class="code-inline">AdaBoostClassifier</span> and <span class="code-inline">GradientBoostingClassifier</span>.
    </p>
    <p>
        Let's test them on the <strong>Moons Dataset</strong> (from the previous chapter) to compare their decision boundaries.
    </p>
</div>

```python
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

# 1. Generate Data (Moons)
X, y = make_moons(n_samples=500, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# 2. Train Models
# AdaBoost: Note that 'SAMME.R' was removed in recent sklearn versions. We use 'SAMME'.
ada_clf = AdaBoostClassifier(
    n_estimators=200, 
    algorithm="SAMME",  # <--- UPDATED HERE
    learning_rate=0.5, 
    random_state=42
)

# Gradient Boosting: The powerful generalizer
gb_clf = GradientBoostingClassifier(
    n_estimators=200, 
    learning_rate=0.1, 
    max_depth=3, 
    random_state=42
)

ada_clf.fit(X_train, y_train)
gb_clf.fit(X_train, y_train)

# --- Plotting Function (Reused) ---
def plot_decision_boundary(clf, X, y, title, ax):
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolor='k', s=20)
    ax.set_title(title)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

plot_decision_boundary(ada_clf, X, y, f"AdaBoost (Acc: {ada_clf.score(X_test, y_test):.2%})", axes[0])
plot_decision_boundary(gb_clf, X, y, f"Gradient Boosting (Acc: {gb_clf.score(X_test, y_test):.2%})", axes[1])

plt.show()
```

<div class="note-box">
    <h2>5. Conclusion: The Power of Iteration</h2>
    <p>
        Let's analyze the decision boundaries we just generated. Both models performed very well (~88-89%), effectively solving the "Moons" problem.
    </p>

<h3>1. AdaBoost (Accuracy: 89.60%)</h3>
    <p>
        Look at the plot on the left.
    </p>
    <ul>
        <li><strong>Visuals:</strong> The boundary is very "blocky" and composed of simple vertical and horizontal lines.</li>
        <li><strong>Why?</strong> AdaBoost typically uses <strong>Decision Stumps</strong> (trees with only 1 split). It builds a complex model by summing up hundreds of very simple, weak opinions. It screams: <em>"This specific region is definitely Blue!"</em> and then adjusts.</li>
    </ul>

<h3>2. Gradient Boosting (Accuracy: 88.00%)</h3>
    <p>
        Look at the plot on the right.
    </p>
    <ul>
        <li><strong>Visuals:</strong> The boundary is slightly more complex and nuanced than AdaBoost.</li>
        <li><strong>Why?</strong> It uses deeper trees (depth=3) that learn to predict the <strong>Residuals</strong> (errors) of the previous trees. It gently "nudges" the decision boundary closer to the truth with every step.</li>
    </ul>

<h3>Bagging vs. Boosting: The Final Verdict</h3>
    <ul>
        <li><strong>Bagging (Random Forest):</strong> Good for reducing variance (smoothing things out). Hard to overfit. Parallel training (Fast).</li>
        <li><strong>Boosting (Ada/GBM):</strong> Good for reducing bias (learning hard patterns). Can overfit if not careful. Sequential training (Slower).</li>
    </ul>
